```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dress_review.csv')
df.head()
```

## Removing unnecessary features

```python
df = df.drop(['s.no', 'clothing_id'], axis=1)
df.head()
```

```python
# Setting Fonts with Different Colors and Families
xfont = {
    'family': 'sans-serif',  # Change the font family to a sans-serif style
    'color': 'Black',         # Change the font color to blue
    'weight': 'normal',     # Use a normal font weight
    'size': 12,
}
yfont = {
    'family': 'sans-serif',
    'color': 'green',        # Change the font color to green
    'weight': 'normal',
    'size': 12,
}
title_font = {
    'family': 'monospace',   # Change the font family to a monospace style
    'color': 'red',          # Change the font color to red
    'weight': 'bold',       # Use a bold font weight
    'size': 14,
    'style': 'italic',      # Apply italic style
}
legend_prop = {
    'weight': 'normal',     # Use a normal font weight
    'size': 14,             # Increase the font size
    'style': 'oblique',     # Apply oblique style
}
highlight = {
    'family': 'cursive',     # Change the font family to a cursive style
    'color': 'darkorange',  # Change the font color to dark orange
    'weight': 'bold',
    'size': 14,             # Increase the font size
}
```

```python
# Remove the trailing space from the column name
df.rename(columns={'recommend_index ': 'recommend_index'}, inplace=True)

# Grouping the data by 'rating' and 'recommend_index'
grouped = df.groupby(['rating', 'recommend_index']).size().unstack()


# Create the paired bar plot
ax=grouped.plot(kind='bar', stacked=True)

# Adding labels and title

ax.set_xlabel('Rating and Recommend Index',fontdict=xfont)

ax.set_ylabel('Count', fontdict=yfont, labelpad=0, loc='center')

# Setting Title
ax.set_title('Stacked Bar Plot: Rating vs. Recommend Index', fontdict=title_font)

# Show the plot
plt.show()
```

```python


# Create a figure with a specific size
fig, ax = plt.subplots(figsize=(12, 8))

# Remove the trailing space from the column name
df.rename(columns={'recommend_index ': 'recommend_index'}, inplace=True)

# Grouping the data by 'rating' and 'recommend_index'
grouped = df.groupby(['rating', 'recommend_index']).size().unstack()

# Define a custom color palette
colors = ['Red', 'Green']  # Customize the colors as needed

# Define the bar width
bar_width = 0.8  # Increase the bar width as needed

# Create the paired bar plot with the custom color palette and bar width
grouped.plot(kind='bar', stacked=True, ax=ax, color=colors, width=bar_width)

# Adding labels and title
ax.set_xlabel('Rating and Recommend Index', fontdict=xfont)
ax.set_ylabel('Count', fontdict=yfont, labelpad=0, loc='center')

# Setting Title
ax.set_title('Stacked Bar Plot: Rating vs. Recommend Index', fontdict=title_font)

# Add values to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=9, color='black')

# Show the plot
plt.show()


```
