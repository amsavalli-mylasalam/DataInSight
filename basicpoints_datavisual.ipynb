One fundamental principle of data visualization is to keep it simple and use visual elements that effectively convey the information.

# Basic Point 1: Use Clear and Concise Labels

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python


# Sample data (you can replace this with your dataset)
varieties = ["Variety 1", "Variety 2", "Variety 3", "Variety 4", "Variety 5"]
modal_prices = [1200, 1200, 1000, 1200, 1200]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(varieties, modal_prices, color='skyblue')
plt.xlabel("Variety of Commodities")
plt.ylabel("Modal Price")
plt.title("Modal Prices for Different Varieties of Commodities")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability

# Display the values on top of the bars
for i, price in enumerate(modal_prices):
    plt.text(i, price, str(price), ha='center', va='bottom')

plt.tight_layout()

# Show the plot
plt.show()
```
